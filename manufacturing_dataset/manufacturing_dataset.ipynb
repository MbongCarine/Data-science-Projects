# **Predicting Defect Rates in Manufacturing**

## **Introduction**  
In the modern manufacturing industry, ensuring product quality is paramount to maintaining operational efficiency, customer satisfaction, and cost-effectiveness. However, defects in products during production can severely impact the overall productivity and profitability of a manufacturing unit. Understanding the factors influencing defect rates is essential to identify potential areas of improvement and optimize production processes. 

## **Objective**  
This project aims to predict defect occurrences using production, supply chain, quality control, maintenance, inventory, workforce, energy, and additive manufacturing data. It does this by developing a predictive model to classify products as **Low Defects (0) or High Defects (1)**, helping manufacturers optimize processes and reduce defects.  

## **Key Features**  
- **Production**: Volume, Cost  
- **Supply Chain**: Supplier Quality, Delivery Delays  
- **Quality Control**: Defect Rate, Quality Score  
- **Maintenance**: Hours Spent, Downtime Percentage  
- **Inventory**: Turnover, Stockout Rate  
- **Workforce & Safety**: Productivity, Safety Incidents  
- **Energy**: Consumption, Efficiency  
- **Additive Manufacturing**: Process Time, Material Cost

**4-Model Splitting, Training and Evaluation**

**4.1-Model Splitting an Scaling**

````python
# Define features (X) and target variable (y)
X = df.drop(columns=["DefectStatus"])  # Drop the target column
y = df["DefectStatus"]  # Target variable
````

````python
# Normalize numerical features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)  # Scaled features

# Convert X_scaled back to DataFrame with original column names
X_scaled_df = pd.DataFrame(X_scaled, columns=X.columns)

# Split dataset into training and testing sets (80% train, 20% test)
X_train, X_test, y_train, y_test = train_test_split(X_scaled_df, y, test_size=0.2, random_state=4)
````

**4.2-Model Training and Evaluation**

````python
# Define models
models = {
    "Logistic Regression": LogisticRegression(),
    "Decision Tree": DecisionTreeClassifier(),
    "Random Forest": RandomForestClassifier(),
    "Gradient Boosting": GradientBoostingClassifier(),
    "SVM": SVC()
}
````

````python
# Dictionary to store model performance
model_performance_no_oversampling = {}

# Train and evaluate models
best_model_no_os = None
best_accuracy_no_os = 0

for name, model in models.items():
    model.fit(X_train, y_train)  # Train model without oversampling
    y_pred = model.predict(X_test)  # Make predictions
    accuracy = accuracy_score(y_test, y_pred)  # Compute accuracy

    print(f"Model: {name}")
    print(f"Accuracy: {accuracy:.4f}")
    print("Classification Report:\n", classification_report(y_test, y_pred))
    print("=" * 50)

    # Store model accuracy
    model_performance_no_oversampling[name] = accuracy

    # Save best model (without oversampling)
    if accuracy > best_accuracy_no_os:
        best_accuracy_no_os = accuracy
        best_model_no_os = model
        best_model_name_no_os = name

    # Plot confusion matrix
    cm = confusion_matrix(y_test, y_pred)
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=["Low Defects", "High Defects"], yticklabels=["Low Defects", "High Defects"])
    plt.xlabel("Predicted Label")
    plt.ylabel("True Label")
    plt.title(f"Confusion Matrix - {name}")
    plt.show()
````

**4.3-Best Model with no oversampling**

````python
# Save the best model (without oversampling)
print(f"Best Model (No Oversampling): {best_model_name_no_os} with Accuracy: {best_accuracy_no_os:.4f}")
joblib.dump(best_model_no_os, "best_defect_model_no_os.pkl")  # Save model

# Load best model
best_model_no_os = joblib.load("best_defect_model_no_os.pkl")

# Compute ROC curve and AUC
y_prob_best_no_os = best_model_no_os.predict_proba(X_test)[:, 1]  # Get probability estimates
fpr_no_os, tpr_no_os, _ = roc_curve(y_test, y_prob_best_no_os)
roc_auc_no_os = auc(fpr_no_os, tpr_no_os)

# Plot ROC curve (No Oversampling)
plt.figure(figsize=(8, 6))
plt.plot(fpr_no_os, tpr_no_os, label=f"{best_model_name_no_os} (AUC = {roc_auc_no_os:.3f})", color="red")
plt.plot([0, 1], [0, 1], linestyle="dotted", color="gray")  # Reference line

plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve - Best Model (No Oversampling)")
plt.legend()
plt.grid(True)
plt.show()
````

From the above plot we can see that training with the imbalanced dataset leads to random forest being the best model with an accuracy of 95.99 and an AUC of 86.7%. It classified 23 high defects as low defects and 7 low defects as high defects.

**5-Oversampling for Balancing dataset**

````python
# Apply SMOTE to balance the dataset
smote = SMOTE(sampling_strategy="auto", random_state=42)
X_resampled, y_resampled = smote.fit_resample(X_scaled_df, y)

# Check class distribution
before_counts = Counter(y)
after_counts = Counter(y_resampled)

print("Before SMOTE:", before_counts)
print("After SMOTE:", after_counts)
````

**5.2-Visuluazing plots of Before sampling and after sampling**

````python
# Convert to DataFrame for plotting
df_counts = pd.DataFrame({
    "Class": ["Low Defects", "High Defects"],
    "Before SMOTE": [before_counts[0], before_counts[1]],
    "After SMOTE": [after_counts[0], after_counts[1]]
})

# Plot class distributions
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Before SMOTE
sns.barplot(x=df_counts["Class"], y=df_counts["Before SMOTE"], palette="pastel", ax=axes[0])
axes[0].set_title("Class Distribution Before SMOTE")
axes[0].set_ylabel("Count")

# After SMOTE
sns.barplot(x=df_counts["Class"], y=df_counts["After SMOTE"], palette="muted", ax=axes[1])
axes[1].set_title("Class Distribution After SMOTE")
axes[1].set_ylabel("Count")

plt.tight_layout()
plt.show()
````

**5.3-Splitting oversampled data to training and testing**

````python
# Split dataset into training and testing sets
X_train1, X_test1, y_train1, y_test1 = train_test_split(X_resampled, y_resampled, test_size=0.2, random_state=42)
````

**5.4-Model Training and Evaluation**

````python
# Define models
models = {
    "Logistic Regression": LogisticRegression(),
    "Decision Tree": DecisionTreeClassifier(),
    "Random Forest": RandomForestClassifier(),
    "Gradient Boosting": GradientBoostingClassifier(),
    "SVM": SVC(),
    "CatBoost": CatBoostClassifier(iterations=500, learning_rate=0.05, depth=6, verbose=0)  # Silent mode
}
````

````python
# Dictionary to store model performance
model_performance = {}

# Train and evaluate models
best_model = None
best_accuracy = 0

for name, model in models.items():
    model.fit(X_train1, y_train1)
    y_pred1 = model.predict(X_test1)
    accuracy = accuracy_score(y_test1, y_pred1)

    print(f"Model: {name}")
    print(f"Accuracy: {accuracy:.4f}")
    print("Classification Report:\n", classification_report(y_test1, y_pred1))
    print("=" * 50)

    # Store model accuracy
    model_performance[name] = accuracy

    # Save best model
    if accuracy > best_accuracy:
        best_accuracy = accuracy
        best_model = model
        best_model_name = name

    # Plot confusion matrix
    cm = confusion_matrix(y_test1, y_pred1)
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=["Low Defects", "High Defects"], yticklabels=["Low Defects", "High Defects"])
    plt.xlabel("Predicted Label")
    plt.ylabel("True Label")
    plt.title(f"Confusion Matrix - {name}")
    plt.show()

# Save the best model
print(f"Best Model: {best_model_name} with Accuracy: {best_accuracy:.4f}")
joblib.dump(best_model, "best_defect_model.pkl")  # Save model to file
````

````python
# Load best model
best_model = joblib.load("best_defect_model.pkl")

# Compute ROC curve and AUC
y_prob_best = best_model.predict_proba(X_test1)[:, 1]  # Get probability estimates
fpr, tpr, _ = roc_curve(y_test1, y_prob_best)
roc_auc = auc(fpr, tpr)

# Plot ROC curve
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, label=f"{best_model_name} (AUC = {roc_auc:.3f})", color="blue")
plt.plot([0, 1], [0, 1], linestyle="dotted", color="gray")  # Reference line

plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve - Best Model")
plt.legend()
plt.grid(True)
plt.show()
````

**Comparison of Model Performance: Before vs. After SMOTE**

**Before SMOTE (Imbalanced Data):**

* **Class Imbalance:** Significant imbalance favoring "High Defects" (1).
* **Model Bias:** Models heavily biased towards predicting "High Defects."
* **Low "Low Defects" (0) Recall:** Models struggled to identify "Low Defects" cases.
* **Misleading Accuracy:** High accuracy scores were misleading due to the class imbalance.
* **Performance:**
    * Logistic Regression: Accuracy 0.8765. Poor "Low Defects" recall.
    * Decision Tree: Accuracy 0.9090. Some improvement in "Low Defects" recall.
    * Random Forest: Accuracy 0.9537. High "High Defects" recall, low "Low Defects" recall.
    * Gradient Boosting: Accuracy 0.9522. Similar to Random Forest.
    * SVM: Accuracy 0.8997. Poor "Low Defects" recall.

**After SMOTE (Balanced Data):**

* **Class Balance:** Perfectly balanced dataset after applying SMOTE.
* **Improved Learning:** Models learned patterns from both classes effectively.
* **Improved "Low Defects" (0) Detection:** Significantly improved recall and precision for "Low Defects."
* **Reliable Metrics:** Accuracy and other metrics became reliable indicators of performance.
* **Performance:**
    * Logistic Regression: Accuracy 0.7917. Performance decreased.
    * Decision Tree: Accuracy 0.9183. Improved performance.
    * Random Forest: Accuracy 0.9495. Excellent balance of precision and recall.
    * Gradient Boosting: Accuracy 0.9468. Excellent balance of precision and recall.
    * SVM: Accuracy 0.8991. Improved performance.
    * CatBoost: Accuracy 0.9633. Best performing model, excellent precision and recall.

**Key Observations:**

* SMOTE effectively balanced the dataset.
* Models showed a significant improvement in identifying "Low Defects" after SMOTE.
* CatBoost emerged as the best performing model.
